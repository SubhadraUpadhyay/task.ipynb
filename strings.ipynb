# wap to take user input and check the data, if the data is numeric,convert it to number else if the data is alphanumeric then extract only number and convert it

```python
data = input("please, enter some data:")
if data.isnumeric():
    num = int(data)
    print ("data converted to numbet:",num)
elif data.isalnum():
    temp = ''
    for char in data:
        if char.isnumeric():
            temp += char
    if temp:
        num = int(temp)
        print("data converted to number:",num)
    else:
        print("your data dose not contain any numbers")
    
```

# wap a simple text menu based calculator for doing basic options like addition, substraction, muitiplication

```python
print("here is boring calculator menu")
print("1. add number")
print("2. subtstract number")
print("3. multiply number")
ch = input ("please, enter your choice:")
if ch == '1':
    num1 = int(input("enter first number"))
    num2 = int(input("enter second number"))
    print("sum is:",num1+num2)
elif ch == '2':
    num1 = int(input("enter first number"))
    num2 = int(input("enter second number"))
    print("difference is:",num1-num2)
elif ch == '3':
    num1 = int(input("enter first number"))
    num2 = int(input("enter second number"))
    print("product is:",num1*num2)
else:
    print("invalid choice")
```
